```python
mj_df_1["state"].value_counts()
```

```python
grouped_data=mj_df_1.groupby("state")
grouped_data.count()
```

```python
grouped_data["state"].unique()
```

```python
df_1=pd.DataFrame({
    "States": grouped_data["state"].unique(),
    "# Mass Shooting":grouped_data["state"].count()
})
df_1.head()
```

```python
df_1=df_1.sort_values('# Mass Shooting', ascending = False)
df_2=df_1[:5].copy()

new_row = pd.DataFrame(data = {
    'States' : ['Others'],
    '# Mass Shooting' : [df_1['# Mass Shooting'][5:].sum()]
})

df_2 = pd.concat([df_2, new_row])
df_2
```

```python
plt.pie(df_2["# Mass Shooting"],labels=df_2["States"],autopct="%1.1f%%",shadow=True, startangle=60)
plt.title("Last Five Years Incidents")
plt.savefig("PercentangeStates.png")
```
